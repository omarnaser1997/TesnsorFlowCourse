```python
import numpy as np 
import matplotlib.pyplot as plt 
import cv2
import tensorflow.keras as k
import tensorflow as tf
```

```python
image = cv2.imread("/Users/omarnasser/StudioProjects/TesnsorFlowCourse/generatedatasets/face.jpeg")
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image) 
```

```python
data_augmentation = k.Sequential(
    [
    k.layers.RandomFlip("horizontal",(128,128,3)),
    k.layers.RandomRotation(0.1),
    k.layers.RandomZoom(0.1)
    ]
)
```

```python

for i in range(25):
    plt.subplot(5,5,i+1)
    img = data_augmentation(tf.expand_dims(image,0))[0]/255
    plt.imshow(img)
    plt.axis(False)
```
