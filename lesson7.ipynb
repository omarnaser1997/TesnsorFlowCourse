```python
from imutils import paths
import cv2
import matplotlib.pyplot as plt
import random
```

```python
# Load the dataset
imagePaths = list(paths.list_images("/Users/omarnasser/Desktop/images"))
random.shuffle(imagePaths)
def getClassNames(path):
    class_names=[]
    for imagePath in path:
        lable = imagePath.split('/')[-3]
        if lable not in class_names:
            class_names.append(imagePath.split('/')[-3])
    return class_names

```

```python
def prepareDataSet(path, labels):
    xTrain=[]
    xTest=[]

    yTrain=[]
    yTest=[]
    for imagePath in path:
        image = cv2.imread(imagePath)
        if 'test' == imagePath.split('/')[-2]:
            xTest.append(image)
            yTest.append(labels.index(imagePath.split('/')[-3]))
        elif 'train' == imagePath.split('/')[-2]:
            xTrain.append(image)
            yTrain.append(labels.index(imagePath.split('/')[-3]))  
    return [xTrain,yTrain,xTest,yTest]



```

```python
labels = getClassNames(imagePaths)
[xTrain,yTrain,xTest,yTest] = prepareDataSet(imagePaths,labels)
yTest
```
